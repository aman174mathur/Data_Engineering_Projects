```python
dataframe1  = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}
```

```python
dataframe2  = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']}
```

```python
df1 = pd.DataFrame(dataframe1 , index = [0,1,2,3])
```

```python
df2= pd.DataFrame(dataframe2 , index = [4,5,6,7])
```

```python
frames = [df1,df2]
```

```python
f=df1.merge(df2, how  = "outer" )       
f#outer because joined  not finding  common
```

```python
y=df1.merge(df2, how  = "outer" , on = "Age" )       
y # joining on Age 
```

```python
x=pd.concat([df1,df2] , join = 'outer' ,axis = 0)  
x    
```
